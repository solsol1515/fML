## 벡터와 벡터의 합

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

a = np.array([1,2])
b = np.array([2,1])

c = a + b
print(c)

x = np.array([[1,2],[2,3]])
y = np.array([[1,1],[2,2]])
z = x + y
print(z)
```

```python
# 차트로 확인
plt.xlim(-3, 5)               
plt.ylim(-3, 5)
plt.grid(True)

# 기준점(0,0)에서 각 벡터를 선으로 표시
plt.annotate('',xy=a, xytext=(0,0), arrowprops=dict(facecolor='black'))
plt.annotate('',xy=b, xytext=(0,0), arrowprops=dict(facecolor='black'))
plt.annotate('',xy=c, xytext=(0,0), arrowprops=dict(facecolor='black'))

# 각 벡터에 점을 표시
plt.plot(0,0,'bo',ms=10) 
plt.plot(a[0],a[1],'ro',ms=10) 
plt.plot(b[0],b[1],'ro',ms=10) 
plt.plot(c[0],c[1],'ro',ms=10) 

# a와 c 그리고 b와 c를 연결
plt.plot([a[0],c[0]],[a[1],c[1]], 'r--')
plt.plot([b[0],c[0]],[b[1],c[1]], 'r--')

# 각 선에 글자 표시
plt.text(0.2, 1.2, '$a$', fontdict={'size':15 })
```

#### [결론] 벡터의 합

두 벡터를 평행사변형으로 만들때 해당되는 나머지 점(=선)이 되는 것이 두 벡터의 합이된다

### 기하학적인 또다른 벡터의 합 표현

기존의 a벡터와 b벡터가 있을 때

1. b 벡터를 a벡터가 닿도록 이동
2. a 벡터와 b벡터의 합을 구하기 (피타고라스 정리 이용 : 삼각형 )


```python
plt.annotate('',xy=a, xytext=(0,0), arrowprops=dict(facecolor='blue'))
plt.annotate('',xy=c, xytext=a, arrowprops=dict(facecolor='blue'))  # b 벡터를 a벡터로 시작점으로 이동
plt.annotate('',xy=c, xytext=(0,0), arrowprops=dict(facecolor='red'))

plt.plot(0,0,'go', ms=10)
plt.plot(a[0],a[1],'go', ms=10)
plt.plot(c[0],c[1],'go', ms=10)

plt.text(0.2, 1.2, '$a$', fontdict={'size':15 })
```
