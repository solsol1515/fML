### numpy를 이용한 벡터의 기하학적 의미

+ 벡터를 공간상에서 표현시 점(point) 또는 화살표(arrow)로 표현할 수 있다
    - a = ([1,2]) 벡터 a에서 
    - x자표 1과 y좌표 2의 값으로 점으로 표시할 수 있고
    - 시작점[0,0]에서 해당 a좌표까지를 화살표 선으로 표현할 수 있다


```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
ax = plt.subplot(111) # 1, 1, 1 동일 = 즉, 하나의 플롯

xvalue = np.arange(0.0, 5.0, 0.01)  # 범위는 0부터 5사이
yvalue = np.cos(2*np.pi* xvalue )
    
lineplot = plt.plot(xvalue, yvalue, lw=2)
```

[참고] 차트상에서 주석 처리하는 함수 : annotate(s, xy, xytext,  arrowprops, . . . )
    + s: 주석
    + xy : 화살표시작위치
    + xytext : 주석 텍스트의 시작 위치
    + arrowprops : 화살표의 속성들

```python
# [참고] 그래프 위의 주석처리
plt.annotate('max value', xy=(3,1), xytext=(4, 1.5), arrowprops=dict(facecolor='red'))
plt.ylim(-2, 2)
lineplot = plt.plot(xvalue, yvalue, lw=2)
```

```python
a = np.array([1,3])
plt.plot(0,0, 'ro', ms=10) # k: black, r:red,  P: plus 모양 , o:o모양  ms=maker size
plt.xticks(np.arange(-2, 5)) # x측 범위
plt.yticks(np.arange(-1, 5))
plt.xlim(-2.5, 4.4)
plt.ylim(-0.5, 4.4)
plt.grid(True)

# a 점 찍기
plt.plot(a[0], a[1], 'bo', ms=10) # b : blue, o : o모양

# 0,0 점과 a 점을 선으로 그리면 이것이 벡터이다(?)
plt.annotate('',xy=a, xytext=(0,0), arrowprops=dict(facecolor='black'))

# 선에 글씨추가
plt.text(0.1, 1.5, "$a$", fontdict={'size':20 })

# a 점에 글씨추가
plt.text(1.2, 3.3, '$(1,3)$', fontdict={'size':20})

# 벡터의 평행 이동 - 시작점도 같이 변경
# plt.annotate('', xy=a+[-1, 1], xytext=(0,0), arrowprops=dict(facecolor='gray'))
plt.annotate('', xy=a+[-1, 1], xytext=(-1,1), arrowprops=dict(facecolor='gray'))
plt.text(-0.9, 2.5, '$a$', fontdict={'size':20})
```

#### 벡터의 길이
    `  2차원 벡터 a의 길이는 피타고라스 정리를 이용하여 얻을 수 있는데, 그 값을 벡터의 놈(norm)이라고 한다.
    `  수학표기법 ||a||
    * numpy에서 linalg 서브패키지에 norm() 이용

```python
a = np.array([1,3])
alength = np.linalg.norm(a)
print(alength)  # 벡터의 길이
```
