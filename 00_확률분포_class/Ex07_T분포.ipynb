###  Student T 분포 ( t distribution)

+ 정규분포보다 예측 범위를 넓힌 분포로서 정규분포 곡선보다 완만하게 표현된다
+ 표본의 수가 적을 때 정규분포보다 t 분포를 사용한다
+ 자유도(degree of freedom)가 클수록 정규분포에 수렴한다.

[참고] 확률 분포
- 이산 확률 분포 : 베르누이분포, 이항분포, 다항분포
- 연속 확률 분포 : 정규분포, 균등분포

[참고] 정규분포에서는 데이터가 많을 수록 신뢰도가 높지만 데이타 양이 작으면 신뢰도가 떨어진다


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy as sp
import pandas as pd
```

```python
""" scipy를 이용한 t분포 알아보기 """

xx = np.linspace(-4, 4,100)

for df in [ 1, 3, 5, 10, 20 ]:
    rv = sp.stats.t(df = df)
    plt.plot(xx, rv.pdf(xx), label = ("student-t (dof = %d)" % df))

plt.plot(xx, sp.stats.norm().pdf(xx), label = "Normal", lw=4, alpha = 0.5)
plt.legend()
plt.grid()
plt.show()
```
